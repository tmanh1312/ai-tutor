```python
import json
import pandas as pd
from openai import OpenAI
import os
```

```python
# Our API key here
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY",""))
```

```python
pd.set_option('display.max_columns', None)

df = pd.read_csv('data/combined.csv.bz2')
df = df[df['gender'] == 'men']

is_shorts = (df['parentCategory_displayName'] == 'Shorts')
is_trousers = (df['parentCategory_displayName'] == 'Trousers')
is_joggers = (df['parentCategory_displayName'] == 'Joggers')
is_pants = (df['parentCategory_displayName'] == 'Pants')
is_tights = (df['parentCategory_displayName'] == 'Tights')
is_sweatpants = (df['parentCategory_displayName'] == 'Sweatpants')

df = df[is_shorts | is_trousers | is_joggers | is_pants | is_tights | is_sweatpants]
df.head()
```

```python
df = df[['product_displayName', 'parentCategory_displayName', 'product_defaultParentCategory', 'product_feel', 'product_fit', 
         'product_function', 'product_activity', 'sku_orderInAvailabilityMap', 'sku_colorCodeDesc', 'list_price', 'sale_price', 'sku_size']]
df.head()
```

```python
import random

def random_question():
    # Randomly select a product
    product = df.sample(n=1).iloc[0]
    product_name = product['product_displayName']
    product_color = product['sku_colorCodeDesc']
    activity = random.choice(activities)

    # Construct a prompt using the product's color
    context_1 = f"Given a customer interested in lululemon products suitable for activities like {activity} and prefers the color {product_color},"
    context_2 = "Ask a question of what type of lululemon product to recommend?"  # Focused on types of products, not specific names
    context_3 = "Do not ask about sizing."  # Constraint
    context_4 = "Do not provide any product names or colors in the prompt. The recommendation MUST be focused on lululemon {product_name} with explanations."
    prompt = context_1 + context_2 + context_3 + context_4
    
    # Query the OpenAI Chat API
    response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": prompt},
    ],
    temperature=1,
    max_tokens=256,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
    )
    
    question = (response.choices[0].message.content)
    return question, product_name
```

```python
def get_response():
    # System's prompt
    system_prompt, product_name = random_question()
    print(f'Question: {system_prompt}')
    # User's prompt
    user_input = input(f'Answer the following question: {system_prompt}')
    print(f'Your answer: {user_input}')

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_input}
        ],
        temperature=1,
        max_tokens=256,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )
    print(response.choices[0].message.content.strip())
    return (response.choices[0].message.content.strip()), product_name
```

```python
def extract_keywords(product):
    # Prepare a list of expected keywords or phrases from product details
    keywords = []
    for col in df.columns:
        value = product[col]
        if isinstance(value, str):
            keywords += value.lower().split(", ")
        elif isinstance(value, (int, float)):
            keywords.append(str(value))
    return set(keywords)
```

```python
def assess_response(user_response, product_name):
    product_details_df = df[df['product_displayName'] == product_name]
    if product_details_df.empty:
        return f"No product found with the name '{product_name}'. Please check the product name and try again."

    product_details = product_details_df.iloc[0]
    keywords = set(str(value).lower() for col, value in product_details.items() if isinstance(value, str))
    if all(keyword in user_response.lower() for keyword in keywords):
        return f"Response for {product_name}: Passes! Successfully covered key aspects."
    else:
        missing_keywords = [keyword for keyword in keywords if keyword not in user_response.lower()]
        return f"Response for {product_name}: Could be improved. Missing details: {', '.join(missing_keywords)}"
```

```python
if __name__ == "__main__":
    responses = []

    # Collect responses 
    for i in range(2):
        user_response, product_name = get_response()
        responses.append((user_response, product_name))

    # Assess responses all at once
    combined_feedback = []
    for response, product_name in responses:
        feedback = assess_response(response, product_name)
        combined_feedback.append(feedback)

    # Summarize and provide collective feedback
    overall_feedback = "Summary of Feedback:\n" + "\n".join(combined_feedback)
    print(overall_feedback)
```
